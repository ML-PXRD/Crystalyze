```python
from pymatgen.core import Lattice, Structure
from pymatgen.symmetry.groups import SpaceGroup

# Example: Create a structure with space group I4/mmm
space_group = 'I4/mmm'

# Define the lattice parameters compatible with the space group
a = 3.84
b = 3.84
c = 6.28
alpha = 90
beta = 90
gamma = 90

lattice = Lattice.from_parameters(a, b, c, alpha, beta, gamma)

# Define the asymmetric unit
# Let's place an atom at the origin and another in a general position
# These positions should be in fractional coordinates
#generate random coordinates
import random
asymmetric_unit = [{"species": "Si", "coords": [0, 0, 0]},
                   {"species": "O", "coords": [0, random.random(), random.random()]}]

# Use from_spacegroup to construct the structure
structure = Structure.from_spacegroup(space_group, lattice,
                                      [site['species'] for site in asymmetric_unit],
                                      [site['coords'] for site in asymmetric_unit])

print(structure)

#analyze the spacegroup
sg = SpaceGroup(space_group)
print(sg)
```
