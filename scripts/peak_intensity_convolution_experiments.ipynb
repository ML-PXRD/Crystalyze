```python
import torch
import torch.nn as nn

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        
        # Define a 1D convolution layer without padding
        self.conv = nn.Conv1d(in_channels=2, out_channels=1, kernel_size=2, stride=1, padding=0)

    def forward(self, x):
        return self.conv(x)

# Create the model
model = SimpleCNN()

# Manually setting the weights and bias
desired_weights = torch.tensor([[[0.5], [0.5]]]) # Using 0.5 just for demonstration
desired_bias = torch.tensor([0.0])

model.conv.weight.data = desired_weights
model.conv.bias.data = desired_bias

# Example input tensor (2x4)
input_tensor = torch.tensor([[1.0, 1.0, 5.0, 7.0],
                             [2.0, 2.0, 6.0, 8.0]])
input_tensor = input_tensor.unsqueeze(0)  # Add batch dimension: 1x2x4

# Get the output
output = model(input_tensor)
print(output.squeeze())  # will print a 1x4 tensor
```
