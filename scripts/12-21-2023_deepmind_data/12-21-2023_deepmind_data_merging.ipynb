```python
import pandas as pd 
import numpy as np 
import torch 
```

```python
def merge_and_save_data(file_type):
    # Paths for the data
    print("loading in the csvs ...") 
    base_path_mp = "/home/gridsan/tmackey/cdvae/data/mp_20/"
    base_path_dm = "/home/gridsan/tmackey/cdvae/data/mp_20_dm/"

    # Load and merge CSV data
    df_mp = pd.read_csv(f"{base_path_mp}{file_type}.csv")
    df_dm = pd.read_csv(f"{base_path_dm}{file_type}.csv")
    merged_df = pd.concat([df_mp, df_dm])
    
    # Load and merge .pt data
    print("loading in the graphs ...") 
    graphs_mp = torch.load(f"{base_path_mp}{file_type}.pt")
    graphs_dm = torch.load(f"{base_path_dm}{file_type}.pt")
    
    merged_graphs = graphs_mp.copy()  # Make a copy of the first dictionary
    merged_graphs.update(graphs_dm)   # Update with the second dictionary

    # Save merged data
    print("saving data ... ") 
    merged_df.to_csv(f"{base_path_dm}{file_type}.csv", index=False)
    torch.save(merged_graphs, f"{base_path_dm}{file_type}.pt")

    return len(merged_graphs) == (len(graphs_mp) + len(graphs_dm))
```
