```python
import pandas as pd

def save_cif_files(df):
    """
    Saves each entry in the 'cif' column of the dataframe to a separate file.
    The filename is taken from the 'material_id' column.

    :param df: pandas DataFrame with 'cif' and 'material_id' columns
    """
    for index, row in df.iterrows():
        file_name = f"/home/gridsan/tmackey/cdvae/scripts/1-06-2024_augmentation/XRDs/Data/Generated/CIFs_open_access/{row['material_id']}.cif"
        with open(file_name, 'w') as file:
            file.write(row['cif'])

# Example usage:
# Assuming test_df is your DataFrame
# save_cif_files(test_df)
```
